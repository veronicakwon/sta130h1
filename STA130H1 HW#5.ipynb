In hypothesis testing, we use data from a sample (sampled values) to estimate an average value (such as mean) of a population otherwise known as the population parameter. The sample statistic (like a sample mean) will represent the true population parameter as it has strong correlation to it, as getting a true value for a population parameter directly is much more difficult. We compare this sample statistic to the population parameter, which is predicted under the null hypothesis. If sampled values show that the sample statistic is much more different than what the null hypothesis predicted for the population parameter, we reject it and accept the alternative hypothesis which is the negation of the null hypothesis. Otherwise, we fail to reject the null hypothesis. This result will refer to the entire population based on sample data.

In hypothesis testing, sampled data is used to construct a distrubtion of a test statistic. A low p-value indicates a low likelihood or the unlikelihood that we get the observed data under the null hypothesis, thus making our null hypothesis less and less believable. With stronger evidence against the null hypothesis, we are more inclined to reject the null hypothesis.

Problem introduction:

What is the relationship between this project's experiment and the original (Fisher and Bristol)?:
The sample sizes for the two experiments are different: the STA130H1 Experiment has a sample size of 80 students while Fisher and Bristol's had only 1 sample size. In STA130H1, each student conducts only one trial while in Fisher and Bristol's, the one sample size (which was Bristol) was involved in multiple trials of tea-tasting. Also, it appears that Bristol is a connoisseur when it comes to tea, while we cannot say the same for every single person in STA130H1. This shows a difference in the two different sample sizes' ability in identifying what was poured first-the tea or the milk. 

Null Hypothesis and Alternative Hypothesis of project's experiment:

The Null Hypothesis is that the students in STA130H1 randomly guess whether or not milk or tea was poured first. The success rate will be 0.5.

The Alternative Hypothesis is that the students in STA130H1 do not randomly guess whether or not milk or tea was poured first. The success rate will be greater than or less than 0.5.

Informal Description of the Null Hypothesis:
Students in STA130H1 do not actually know how to distinguish whether or not milk or tea was poured in first, they are simply taking a random guess that may be correct or incorrect.


Description of project experiment:
The population is the students of STA130H1.
The sample is the randomly chosen 80 STA130H1 students.
Test statistic: The number of correct answers (49 / 80)

Quantitative analysis:

For this project, I decided to use a binomial distribution to simulate the number of correct answers under the Null Hypothesis since a binomial distribution displays only two outcomes (correct or incorrect). Since the outcomes of this experiment is not a number but rather a category (correct/incorrect), using a test statistic such as median/mode/mean would not be appropiate as those are calculated using numbers. Instead, the test statistic should be the number of correct guesses, and thus will be used to see if we should reject the null hypothesis or not.

Code explanation: 

First the parameters: there are 80 students in total, thus number_of_students_total = 80. The "value" (or observed value) of the null hypothesis is 0.5 (because whether the answer is correct/incorrect, it is based on random guessing), and correct_answers is 49 (our test statistic). We should make sure the value of p_value remains the same on each run since p_value is based on results_of_simulation which is generated randomly. Thus, we use np.random.seed(3). 
 
Second, simulate 100,000 binomial experiments. np.random.binomial incorporates the np.random function to run 100,000 trials with the given number of students and calculate (the random) number of correct guesses for each trial. This number is stored in results_of_simulation.

Third, we then compare the results of our simulation (there are 100,000) and see if they are greater than or equal to 49 correct answers (our test statistic). This is because this boolean counts the number of times the value of results_of_simulation goes beyond our expectation that is based on the given number of correct guesses. This is essentially calculating the probability that we observe our test statistic under the null hypothesis. If we end up rejecting the null hypothesis, it is implied that there are not many instances where the results_of_simulation >= correct_answers. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
number_of_students_total = 80
value_of_H0 = 0.5
correct_answers = 49

# Ensures that result of p_value is consistent each run
np.random.seed(3)

# Simulate 100,000 binomial experiments
results_of_simulation = np.random.binomial(n=number_of_students_total, p=value_of_H0, size=100000)

# Calculate the p-value
p_value = np.mean(results_of_simulation >= correct_answers)

print("p_value:")
print(p_value)

# Plot the distribution of simulated results
plt.hist(results_of_simulation, bins=np.arange(25, 80) - 0.5, edgecolor='black', alpha=0.7)
plt.axvline(correct_answers, color='red', linestyle='dashed', linewidth=2, label=f'Correct = {correct_answers}')
plt.title('Distribution of Simulated Correct Guesses\nUnder Null Hypothesis (Random Guessing)')
plt.xlabel('Number of Correct Guesses')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Conclusion:

It appears that the calculated p-value is 0.02894 (fixed). Comparing this to the alpha significance level (which is default 0.05), the p_value falls much lower than this number. That means this p-value is considered low, showing that the chance of our test statistic being a generated simulation result is also low. Thus, our test statistic is inconsistent with our null hypothesis, and so we reject the null hypothesis which states that the correct guesses are based on pure randomness.
