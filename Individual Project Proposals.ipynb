# Analysis 1:

## Research Question

Is there an association between the amount of time a person under the age of 30 spends on social media per day and their interest in  activities?

## Variables:

- CONNECTION_social_media_time_per_day (predictor/numeric variable) 
- WELLNESS_phq_little_interest (outcome/categorical variable)

Social media can spark motivation or cause jealousy, aligning with CSCS's prompt.

### Visualizations:
We can group the two columns of variables based on age. 
A KDE plot will visualize the distribution of social media time and interest in activities, with the x-axis representing time spent on social media and the y-axis representing density. There'll be several curves which are categorized as levels of interest in activities.



## Analysis:

We can perform linear regression to examine a linear correlation between these variables assuming residuals are normally distributed,  independent, homoscedastic and reflect linearity. Using CSCS data and statsmodels.formula.api, we'll fit a linear model and obtain summary statistics including the slope. Through bootstrapping, we'll construct a 95% confidence interval for the slope estimate.

## Hypotheses

Null Hypothesis: There is no linear association between daily social media time and level of interest in activity.
Alternative Hypothesis: There is a linear association between daily social media time and level of interest in activity.

Possible results: 

95% CI includes 0 &rarr; no linear association,
95% CI excludes 0 (positive) &rarr; positive correlation,
95% CI excludes 0 (negative) &rarr; negative correlation


# Analysis 2:

## Research Question

Does older age affect how satisfied a person is in their life?

## Variables:

DEMO_age, WELLNESS_subjective_happiness_scale_happy (numeric)

Age may reflect experiences and responsibilities affecting life satisfaction, which is relevant to CSCS's goals.


### Visualizations:
Two histograms (one for ages below 40, another for ages 40 and above) with x-axis and y-axis representing the subjective happiness score and count, respectively, to compare the shapes and distributions.

## Analysis:

Bootstrapping can be used to compare the mean subjective happiness scores in both groups under the assumptions of an independent, representative, and large sample. A 95% confidence interval will have a range of the differences in population means. 

## Hypotheses

Null Hypothesis: Age has no effect on life satisfaction.
Alternative Hypothesis: Age has an effect on life satisfaction.
Possible results:

95% CI includes 0 &rarr; age has no effect on life satisfaction,
95% CI excludes 0 (negative) &rarr; age has a negative effect on life satisfaction,
95% CI excludes 0 (positive) &rarr; age has a positive effect on life satisfaction
